```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/train.csv')
data
```

```python
data.pivot_table('PassengerId', 'Survived', 'Pclass', 'count').plot(kind='bar', stacked=True)
```

```python
fig, axes = plt.subplots(ncols=2)
data.pivot_table('PassengerId', ['SibSp'], 'Survived', 'count').plot(ax=axes[0], title='SibSp')
data.pivot_table('PassengerId', ['Parch'], 'Survived', 'count').plot(ax=axes[1], title='Parch')
```

```python
data.pivot_table('PassengerId', 'Survived', 'Embarked', 'count').plot(kind='bar', stacked=True)
```

```python
data.pivot_table('PassengerId', 'Sex', 'Survived', 'count').plot(kind='bar', stacked=True)
```
